This assignment reads a data file and performs analysis to reach conclusions on the data.

First, we include the libraries needed in this assignment.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
```

Now, we read the Irirs.txt file into a Pandas DataFrame.

```python
iris_df = pd.read_csv('iris.txt')
```

Once the data is read, we check to see if the data is stored correctly by viewing the first and last lines of the data.

```python
n=5
print("The file contains {} entries and {} columns.".format(iris_df.shape[0], iris_df.shape[1]))
print("\nThe first {} rows:\n{}".format(n, iris_df.head(n)))
print("\nThe last {} rows:\n{}\n".format(n, iris_df.tail(n)))

iris_df.info()
```

The Iris dataset has 150 entries and 5 columns. The index is automatically created upon importing of dataset.

There are 4 columns with floating point numbers (the lengths and widths of the petals and sepals) and one colums with text denoting the species name.

Summary statistics on the numerical data:

```python
iris_df.describe()
```

This summary is for all species of plans in the dataset.  It shows that there are no missing values.

Next, we look at some information about the species names.

```python
iris_df.describe(include=['O'])
```

There are 3 different species in the dataset.

The Box & Whiskers chart will show how data in each numerical column is distributed.

```python
iris_df.boxplot(rot="45", return_type='axes')
```

`SepalWidth` appears to have a relatively tight distributoion of values, while `PetalLenth` has a broad range.

Now, we generate line charts for the data.

```python
iris_df.plot(subplots=True, figsize=(12, 10))
```

The Petal charts (width and length) are interesting, in that one could almost distinguish when the data changes species (around the 50 and 100 marks).

Meanwhile, very little can be inferred from the Sepal line charts.

Finally, a histogram of the data is generated to view the frequency of each variable.

```python
iris_df.plot.hist()
```

Now, we will group the data by the "Species" column and view summary statistics on the grouped data.

```python
iris_df.groupby('Species').describe()
```

Now we can start seeing what differentiates the species, as it pertains to petal and sepal size.  For example, it appears that, on average, the Virginica has the largest petals and sepals.

Box plots on the grouped data help in visualizing the above table.

```python
iris_df.boxplot(rot="90", return_type='axes', by='Species')
```

Here we can see the differences between the species.  Note how the `PetalLength` is significantly different for each species.

Finally, a scatter matrix can be constructed to see the relationships between the variables.

```python
scatter_matrix(iris_df, figsize=(10, 10), diagonal='hist', c=["#0392cf", "#7bc043", "#ee4035"])
```

Here, the dirrefent colors represent a species and one could see the relationships between the variables.  For example, there appears to be a very high correlation between `PetalLength` and `PetalWidth`.  This is not the case with `SepalLength` and `SepalWidth`.

Another way is to generate a scatter matrix for each individual species.

```python
for name, group in iris_df.groupby(['Species']):
    scatter_matrix(group, figsize=(10, 10), diagonal='hist')
    plt.suptitle(name)
```

Now we can see the characteristics of the individual species.  

Some interesting things of note:
* Setosa `SepalLength` and `SepalWidth` are very correlated.  These are the only variables for the setosa with a strong correlation.
* Versicolor has a relatively strong correlations of PetalLength and `SepalWidth` to `PetalWidth`.
* Virginica has a high correlation between `PetalLength` and `SepalLength`.
